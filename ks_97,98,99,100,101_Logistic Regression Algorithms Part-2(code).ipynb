```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
data.head(10)
```

#### ekhon Data ke I/O and O/P er modde Devide korbo

```python
real_x = data.iloc[:,[2,3]].values # I/O = IV
real_y = data.iloc[:,4].values     # O/P = DV
```

```python
real_x
```

```python
real_y
```

#### I/O or O/P e Data ke Split korbo

```python
from sklearn.model_selection import train_test_split # aage eta use hotw-->from sklearn.cross_validation import train_test_split
training_x,test_x,training_y,test_y = train_test_split(real_x,real_y,test_size = 0.25, random_state=0)
```

```python
training_x
```

```python
test_x
```

# ks_98_Logistic Regression Algorithms Part-3(code)

### ---> এখন korbo Features Secale 
er kaj holo Data ke ekta Range e ana... -2 theke +2 er vitor e ana(Range).
Bcoz, Age er Basically 10-100 er bitor But Salary holo 10K-100k...
..Huge difference tai Scalling kore hoy...Then Data Ploat korbar Jonno Valid paramrter pabo

```python
from sklearn.preprocessing import StandardScaler 
```

```python
scaler = StandardScaler()
training_x = scaler.fit_transform(training_x) # Scalling
test_x = scaler.fit_transform(test_x)         # Scalling
```

```python
training_x  # ei kane dektesi Data -2 theke +2 er bitore 
```

```python
test_x      # ei kane dektesi Data -2 theke +2 er bitore 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifer_LR = LogisticRegression(random_state=0)
classifer_LR.fit(training_x,training_y)
```

```python
y_pred = classifer_LR.predict(test_x)
y_pred   # ei khane Data 0/1 form e asce  # mne kar kar kase Car Sell kiorte Parbe
```

### Check korbo "y_pred(Prediction)" er Sate "test_y" er Sate

```python
test_y  # Max mile kisu mile nah
```

# ks_99_Logistic Regression Algorithms Part-4(code)

### Confusion Matrix
etar kaj holo "y_pred(Prediction)" er Sate "test_y" value Match check kora == Mane Right and Wrong Prediction amdr Bahir kore Dibe...BUT--->> amra nijera caile koirte parbo but onk Data Hole seta Impossible Hoiya Jai

```python
from sklearn.metrics import confusion_matrix
```

```python
confus_matrix = confusion_matrix(test_y,y_pred)  # y_true = Real Value, y_pred = Prediction er value
confus_matrix
```

# Out of 100 :
TruePostive + TrueNegative = 65 + 24 = 88 (Right).....
FalsePositive + FalseNegative = 3 + 8 =13 (Wrong)

Performance measure – Accuracy
* Accuracy = 100-13 = 87

```python
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(test_y, y_pred))
```

# ks_100_Logistic Regression Algorithms Part-5(code)  --> Video Abar Dekte hbe

### Training Data ke Ploat korbo

```python
from matplotlib.colors import ListedColormap
X_set, y_set = training_x,training_y
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1,  
                               stop = X_set[:, 0].max() + 1, step = 0.01), 
                     np.arange(start = X_set[:, 1].min() - 1,  
                               stop = X_set[:, 1].max() + 1, step = 0.01)) 
  
plt.contourf(X1, X2, classifer_LR.predict( 
             np.array([X1.ravel(), X2.ravel()]).T).reshape( 
             X1.shape), alpha = 0.75, cmap = ListedColormap(('purple','green'))) 
  
plt.xlim(X1.min(), X1.max()) 
plt.ylim(X2.min(), X2.max()) 
  
for i, j in enumerate(np.unique(y_set)): 
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], 
                c = ListedColormap(('red', 'green'))(i), label = j) 
      
plt.title('Classifier (Traning set)') 
plt.xlabel('Age') 
plt.ylabel('Estimated Salary') 
plt.legend() 
plt.show() 
```
